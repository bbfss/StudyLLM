# Tokenizer 基本使用

Autotokenizer 会根据导入的模型自动选择分词器

# 加载和保存

```python
from transformers import AutoTokenizer 
```

```python
sen = "你好世界"
```

```python
# huggingface会根据我们导入的模型来选择对应的tokenizer,这里是直接从网上下载的bert-base-chinese的tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-chinese")
tokenizer
```

```python
# tokenizer保存到本地

tokenizer.save_pretrained("./tokenizer")
```

```python
# 也可以直接从本地加载tokenizer
tokenizer = AutoTokenizer.from_pretrained("./tokenizer")
tokenizer
```

# 句子分词

```python
# tokenizer进行分词

token = tokenizer.tokenize(sen)
token 
```

# 查看字典

```python
tokenizer.vocab
```

```python
tokenizer.vocab_size
```

# 索引转换

```python
ids = tokenizer.convert_tokens_to_ids(token)
ids
```

```python
tokens = tokenizer.convert_ids_to_tokens(ids)
tokens
```

## 更加简便的方式

```python
ids = tokenizer.encode(sen)
ids
```

```python
# 这里添加了[CLS]和[SEP]两个特殊符号用来区分句子的开始和结束
tokens = tokenizer.decode(ids)
tokens 
```

# 填充和截断

```python
ids = tokenizer.encode(sen, padding='max_length', max_length=15)
ids
```

```python
ids = tokenizer.encode(sen, max_length=6, truncation=True)
ids
```

```python
# 分词的同时把attentionmask也生成
inputs = tokenizer(sen, padding="max_length", max_length=15)
inputs
```

# 处理batch数据

```python
sens = ["弱小的我也有大梦想",
        "有梦想谁都了不起",
        "追逐梦想的心，比梦想本身，更可贵"]
res = tokenizer(sens)
res
```
