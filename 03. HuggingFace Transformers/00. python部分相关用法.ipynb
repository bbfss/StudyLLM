# 保存模型（需要初始化）

这个方法会保存模型的参数, 但是仍然初始化模型的结构， 才能加载 

```python
import torch 
import torchvision.models as models

model = models.vgg16(pretrained=True)
torch.save(model.state_dict(), "vgg16_model_weights.pth")
```

```python
model = models.vgg16()
model.load_state_dict(torch.load('vgg16_model_weights.pth'))
model.eval()
```

# 保存模型（无须初始化）

这个方法会把模型的结构也保存进去， 不需要再进行初始化了

```python
import torch 
import torchvision.models as models

model = models.vgg16(pretrained=True)
torch.save(model, "vgg16_model_weights.pth")
```

```python
model = torch.load('vgg16_model_weights.pth')
```
