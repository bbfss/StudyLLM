# 下载actor的基础模型facbook/opt-1.3b

```python
import os

# 1. 基础环境配置
# 设置国内镜像站地址，这是目前最稳的下载方式
os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'

# 强制关闭可能会导致“0 B/s”卡死的极速插件（稳定性优先）
os.environ['HF_HUB_ENABLE_HF_TRANSFER'] = '0'

# 2. 清除 AutoDL 默认代理（防止流量绕路导致镜像站访问超时）
os.environ.pop('http_proxy', None)
os.environ.pop('https_proxy', None)
```

```python
from huggingface_hub import snapshot_download

# 3. 配置下载参数
repo_id = 'facebook/opt-1.3b'
local_dir = '/root/autodl-tmp/StudyLLM/02. Deepseed/opt-1.3b'

# 自动创建目标文件夹（防止路径不存在报错）
if not os.path.exists(local_dir):
    os.makedirs(local_dir, exist_ok=True)

print(f"正在准备下载模型: {repo_id}")
print(f"目标保存路径: {local_dir}")
print("-" * 30)


    # 4. 开始下载
snapshot_download(
    repo_id=repo_id,
    local_dir=local_dir,
    local_dir_use_symlinks=False,  # 不使用符号链接，直接下载文件到本地
    resume_download=True,          # 支持断点续传
    token=None                     # 公开模型不需要 Token
)
print("-" * 30)
print("✅ 下载完成！你可以开始后续的实验了。")
```
